# Исследование закономерностей об успешности игр.

Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Перед вами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и вы планируете кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируете ли вы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

#### Преобразование типов данных

Преобразуем данные в нужные типы.

Посмотрим на уникальные значения в столбце `user_score`.

```python
data['user_score'].unique()
```

Помимо пропущенных значений, можем увидеть аномальное значение `tbd`, что скорее всего означает аббривиатуру `To Be Determined`. Предположим, что это означает, что недостаточно отзывов для формирования рейтинга, либо на момент выгрузки рейтинг не был сформирован.

```python
data.query('user_score =="tbd"')
```

Так как нам нужно привести этот столбец к числовому типу, заменим `tbd` и `NaN` и приведем к числовому типу.

```python
data['user_score']=data['user_score'].replace('tbd', np.NaN)
```

```python
data['user_score']=pd.to_numeric (data['user_score'],errors='coerce')
```

```python
data.info()
```

### Проверка гипотез.

#### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

H_0: Средний пользовательский рейтинг (user_score) платформы "XOne" = Средний пользовательский рейтинг (user_score) платформы "PC"

H_a: Средний пользовательский рейтинг (user_score) платформы "XOne" ≠ Средний пользовательский рейтинг (user_score) платформы "PC"

alpha = 0.01

```python
rating_XOne = data[(data['platform'] =='XOne')& (data['user_score'].notna())] ['user_score']
rating_XOne.describe()
```

```python
rating_PC = data[(data['platform'] =='PC')& (data['user_score'].notna())] ['user_score']
rating_PC.describe()
```

```python
alpha = 0.01 
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(rating_XOne, rating_PC)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Значение p-value показывает: вероятность случайно получить такое или большее различие,при условии, что нулевая гипотеза верна, равно почти 10%. Можно сделать вывод о том, что средние рейтингов в целом похожи друг на друга, что и подтверждает нулевая гипотеза.

#### Средние пользовательские рейтинги жанров Action  и Sports  разные.

H_0: Средний пользовательский рейтинг (user_score) жанра "Action" = Средний пользовательский рейтинг (user_score) жанра "Sports"

H_a: Средний пользовательский рейтинг (user_score) платформы "Action" =≠ Средний пользовательский рейтинг (user_score) платформы "Sports"

alpha = 0.01

```python
genre_action= data[(data['genre'] =='Action')& (data['user_score'].notna())] ['user_score']
genre_action.describe()
```

```python
genre_sports= data[(data['genre'] =='Sports')& (data['user_score'].notna())] ['user_score']
genre_sports.describe()
```

```python
alpha = 0.01 
# если p-value окажется меньше него - отвергнем гипотезу

results = st.ttest_ind(genre_action, genre_sports)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу') 
```

Тест показал, что мы можем отвергнуть нулевую гипотезу о том, что средние пользовательские рейтинги жанров "Action" и "Sports" одинаковые. Мы можем говорить о том, что значения различны.

Для проверки гипотез мы применяли двухвыборочный t-критерий, который используется для определения того, равны ли две средние значения генеральной совокупности .

Чтобы принять или отвергнуть нулевую гипотезу, мы посчитали вероятность получить наблюдаемое на выборке значение при условии, что нулевая гипотеза верна. Если вероятность большая,как в первом случае, мы не можем отвергать нулевую. Во втором же случае вероятность была маленькая, что позволило нам отвергнуть нулевую гипотезу.
